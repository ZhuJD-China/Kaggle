````python
print("Use polynomial features")
from sklearn import preprocessing

poly = preprocessing.PolynomialFeatures(degree=2) # Generate polynomial and interaction features
"""
degreeinteger
    The degree of the polynomial features. Default = 2.
例如：有 a、b 两个特征，那么它的 2 次多项式的次数为 [1,a,b,a^2,a*b,b^2]。
"""
features_ = poly.fit_transform(features)
"""
    C : float, default=1.0
        Inverse of regularization strength; must be a positive float.
        Like in support vector machines, smaller values specify stronger
        regularization.
"""
clf = LogisticRegression(C=20,solver='lbfgs',max_iter=9000)
clf.fit(features_, target)
print(clf.score(features_, target))
````

````python
# print("Use polynomial features")
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
import xgboost as xgb
from xgboost import XGBClassifier
import shap

# load JS visualization code to notebook
shap.initjs()

train = pd.read_csv("./train.csv")
test = pd.read_csv("./test.csv")
clean_data(train)
clean_data(test)

train_features_xgboost = train[["Pclass", "Age", "Sex", "Fare", "SibSp", "Parch", "Embarked"]].values
test_features_xgboost = test[["Pclass", "Age", "Sex", "Fare", "SibSp", "Parch", "Embarked"]].values
target = train["Survived"].values

# poly = preprocessing.PolynomialFeatures(degree=2) # Generate polynomial and interaction features
# polynomial_features = poly.fit_transform(features)

# train_features = poly.fit_transform(train_features_xgboost)
# test_features = poly.fit_transform(test_features_xgboost)

train_features = train_features_xgboost
test_features = test_features_xgboost


xg = xgb.XGBClassifier(
    n_estimators=3000, # 总共迭代的次数，即决策树的个数
    min_child_weight=0.1,
    learning_rate=0.001,
    max_depth=7,  # 树的深度
    subsample=0.47,  # 训练每棵树时，使用的数据占全部训练集的比例
    colsample_bytree=0.35,   # 训练每棵树时，使用的特征占全部特征的比例
    gamma=0.35,   # 惩罚项系数，指定节点分裂所需的最小损失函数下降值
    reg_lambda=0.4,
    random_state=42,
    n_jobs=-1,
)

xg.fit(train_features, target, eval_metric='auc')

prediction_forest = xg.predict(test_features)
write_prediction(prediction_forest, "./xgboost.csv")

scores = model_selection.cross_val_score(xg, train_features, target, scoring='accuracy', cv=10)
print(scores)
print("mean:",scores.mean())
````

## Neural Network

````python
data_train = pd.read_csv('./train.csv')
data_test = pd.read_csv('./test.csv')
clean_data(data_train)
clean_data(data_test)

train_features_ = data_train[["Pclass", "Age", "Sex", "Fare", "SibSp", "Parch", "Embarked"]].values
test_features_ = data_test[["Pclass", "Age", "Sex", "Fare", "SibSp", "Parch", "Embarked"]].values
train_target_ = data_train["Survived"].values

print(train_features_.shape)
print(test_features_.shape)
print(train_target_.shape)
````

````python
import tensorflow as tf
tf.compat.v1.disable_eager_execution()

with tf.compat.v1.Session() as sess:  # 将样本标签转为独热编码的形式
    train_target_ = sess.run(tf.one_hot(train_target_, 2))
print(train_target_.shape)
````

````python
global_step = tf.Variable(0, trainable=False)  # 动态变化的学习轮数
"""  learning_rate,初始学习率
    global_step,当前迭代次数
    decay_steps,衰减速度（在迭代到该次数时学习率衰减为earning_rate * decay_rate）
    decay_rate,学习率衰减系数，通常介于0-1之间。
    staircase=False,(默认值为False,当为True时，（global_step/decay_steps）则被转化为整数) ,选择不同的衰减方式。
    name=None"""
learning_rate = tf.compat.v1.train.exponential_decay(0.60, global_step, 10, 0.96)  # 动态学习速率
````

````python
# 将数据集拆成训练集和测试集
data_tr, data_te, label_tr, label_te = train_test_split(train_features_, train_target_, test_size=0.2)
print(data_tr.shape)
print(data_te.shape)
print(label_tr.shape)
print(label_te.shape)
````

````python

'''
搭建神经网络（定义计算）
'''
x_data = tf.compat.v1.placeholder(tf.float32, [None, 7])  # 占位符：网络的输入
y_data = tf.compat.v1.placeholder(tf.float32, [None, 2])  # 占位符：网络的目标输出

w0 = tf.Variable(tf.zeros([7, 14]))  # 隐层神经元的权值矩阵
w1 = tf.Variable(tf.zeros([14, 2]))  # 输出层神经元的权值矩阵
b0 = tf.Variable(tf.zeros([14]))  # 隐神经元的阈值
b1 = tf.Variable(tf.zeros([2]))  # 输出层神经元的阈值

H = tf.sigmoid(tf.matmul(x_data, w0) + b0)  # 隐层神经元的输出
y = tf.nn.softmax(tf.matmul(H, w1) + b1)  # 网路的输出

cross_entropy = tf.reduce_mean(-tf.reduce_sum(y_data * tf.compat.v1.log(y), axis=1))  # 交叉熵
optimizer = tf.compat.v1.train.GradientDescentOptimizer(learning_rate)
train = optimizer.minimize(cross_entropy, global_step=global_step)  # 训练节点
acc = tf.reduce_mean(tf.cast(tf.equal(tf.argmax(y_data, 1), tf.argmax(y, 1)), dtype=tf.float32))  # 精度

sess = tf.compat.v1.Session()
sess.run(tf.compat.v1.global_variables_initializer())
Learning = []
for i in range(1500):
    sess.run(train, feed_dict={x_data: data_tr, y_data: label_tr})  # 模型训练
    acc_tr = sess.run(acc, feed_dict={x_data: data_tr, y_data: label_tr})
    print(i, 'global_step:', sess.run(global_step), 'acc_tr', acc_tr)
    Learning.append(sess.run(learning_rate))

acc_te = sess.run(acc, feed_dict={x_data: data_te, y_data: label_te})
print('测试样本的精度：', acc_te)
sess.close()
````

````python
plt.plot(Learning)
plt.show()
````
